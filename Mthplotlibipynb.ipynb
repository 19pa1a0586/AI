```
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```
views=[45,76,78,99,87,45,76,38,90] #If we give any single variable then it consider as y-axis
plt.plot(views) #plot means line
plt.show() 
```

```
views=[45,76,78,99,87,45,76,38,90] 
days=range(1,10)
plt.plot(days,views) #plt.plot(x,y)
plt.show() 
```

```
views=[45,76,78,99,87,45,76,38,90] 
days=range(1,10)
plt.plot(days,views,label="Ice creams") #plt.plot(x,y,label)
plt.xlabel("Number of days")
plt.ylabel("Number of Ice creams")
plt.legend(loc="low right")
plt.show() 
```
